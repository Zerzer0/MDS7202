```python
class Profiler():

    def __init__(self, data):
        self.data = data

        # Obtiene la fecha actual
        fecha_actual = datetime.now()

        # Formatea la fecha en el formato DD-MM-YYYY
        fecha_actual = fecha_actual.strftime("%d-%m-%Y")

        ruta = f"EDA_{fecha_actual}"

        if not os.path.exists(ruta):
            os.makedirs(ruta)


        self.dir_fecha=ruta
        self.ft = None
        self.ct = None

    def summarize(self, var):
        #for var in variables:
            print(f"An√°lisis variable {var}")
            print(" ")
            df=self.data[var]
            tipo= str(df.dtype)
            print(f"La columna {var} es de tipo {tipo}")

            print(" ")

            val_unicos= df.unique()
            count_unicos=df.value_counts()
            percent_unicos=(df.value_counts(normalize=True)*100).round(2)
            print(f"Los valores unicos de la columna {var} y sus respectivos porcentajes son: ")
            for i in val_unicos:
                print(f"valor: {i}, Cantidad {count_unicos[i]}, Porcentaje {percent_unicos[i]}%")

            print(" ")

            val_nan= df.isna().sum()
            percent_nan = val_nan/len(df)
            print(f"La variable {var} posee {val_nan} valores nulos ({percent_nan}%)")

            print(" ")

            if tipo == "int64":
                val_cero= (df < 0).sum()
                val_neg = (df == 0).sum()
                
                q1= df.quantile(0.25)
                q3= df.quantile(0.75)
                iqr=q3-q1

                lim_inf= q1-1.5*iqr
                lim_sup= q3+1.5*iqr

                outliers= ((df < lim_inf) | (df > lim_sup)).sum()

                print(f"La variable {var} contiene:")
                print(f"valores cero: {val_cero} ({val_cero/len(df)}%)")
                print(f"valores negativos: {val_neg} ({val_neg/len(df)}%)")
                print(f"Outliers: {outliers} ({outliers/len(df)}%)")

                print(" ")
                print("Otra informaci√≥n descriptiva de la variable:")
                print(df.describe())


                ruta_txt=os.path.join(self.dir_fecha, "summary.txt")
                with open(ruta_txt, 'w', encoding="utf-8") as arch:
                    arch.write(f"An√°lisis variable {var}\n")
                    arch.write("\n")
                    arch.write(f"La columna {var} es de tipo {tipo}\n")
                    
                    arch.write("\n")
                    arch.write("\n")
                    arch.write(f"Los valores unicos de la columna {var} y sus respectivos porcentajes son:\n")
                    for i in val_unicos:
                       arch.write(f"valor: {i}, Cantidad {count_unicos[i]}, Porcentaje {percent_unicos[i]}%\n")
                    arch.write("\n")
                    arch.write(f"La variable {var} posee {val_nan} valores nulos ({percent_nan}%)\n")
                    arch.write("\n")

                    if tipo == "int64":
                    
                        arch.write(f"La variable {var} contiene:\n")
                        arch.write(f"valores cero: {val_cero} ({val_cero/len(df)}%)\n")
                        arch.write(f"valores negativos: {val_neg} ({val_neg/len(df)}%)\n")
                        arch.write(f"Outliers: {outliers} ({outliers/len(df)}%)\n")

                        arch.write("\n")
                        arch.write("Otra informaci√≥n descriptiva de la variable:\n")

                        arch.write(str(df.describe()))
                    


    
    


            

    def plot_vars(self, var, N):
        ruta_txt=os.path.join(self.dir_fecha, "plots")

        if not os.path.exists(ruta_txt):
            os.makedirs(ruta_txt)
        numeric_var=[]
        object_var = []
        
        for v in var:
          df=self.data[v]
          tipo= str(df.dtype)

          if tipo == "object":
             object_var.append(v)

             count=df.value_counts()
             n_most = count.head(N).index.tolist()
             filter= df[df.isin(n_most)]

             fig = px.histogram(filter, title=f'Histograma de variable {v}')
             fig.update_xaxes(title_text=f'{v}')  
             fig.update_yaxes(title_text='Frecuencia')  


             fig.show()









          elif tipo == "int64":
             numeric_var.append(v)


             



             sns.kdeplot(df, shade=True, color='blue')  
             plt.title(f'Distribuci√≥n de Densidad de variable {v}') 
             plt.xlabel('f{v}')  
             plt.ylabel('Densidad')  


             plt.show()
        

        corr=self.data[numeric_var].corr()

        plt.figure(figsize=(10, 8))  # Establece el tama√±o de la figura

        sns.heatmap(corr, annot=True)

        plt.title('Mapa de Calor de Correlaci√≥n')  # Establece el t√≠tulo del gr√°fico
        plt.show()

        

        

        

        

    

    def clean_data(self, variables=None, tecnica="dropna"):
        """Funci√≥n que permite limpiar el dataframe de filas duplicadas y valores nulos, adem√°s de poder separar columnas no at√≥micas.

        Args:
            variables (List, optional): Lista de variables de interes para limpiar. Defaults to None.
            tecnica (str, optional): T√©cnica que se quiere usar para trabajar valores nulos. Defaults to "dropna".

        Raises:
            ValueError: Error que sale cuando se ingresa una t√©cnica no v√°lida.

        Returns:
            .csv: Se guarda archivo .csv con los datos ya limpios.
        """

        # Creaci√≥n de carpeta
        clean_dir= os.path.join(self.dir_fecha, "clean_data")
        os.makedirs(clean_dir, exist_ok=True)

        # Funci√≥n auxiliar
        def limpieza(dataframe, variables=variables, tecnica=tecnica):
            df = dataframe.copy()

            # Filtraci√≥n de variables
            if variables == None:
                variables = df.columns
            variables = list(variables)

            variables_nuevas = []

            # Drop de valores duplicados
            df.drop_duplicates(inplace=True)

            # Separaci√≥n datos no at√≥micos
            for var in variables:
                data = df[var]
                var_type = data.dtype
                if is_object_dtype(var_type):
                    regex_pattern = "[*:()?]+"
                    non_atomic_values = df[var].str.contains(regex_pattern, regex=True)
                    if non_atomic_values.all():
                        # Separa los datos en la columna usando regex
                        columnas = df[var].str.split(regex_pattern, regex=True, expand=True).columns
                        for i in columnas:
                            variables_nuevas.append(f"{var}_{i}")
                        df[variables_nuevas] = df[var].str.split(regex_pattern, regex=True, expand=True)

                        # Eliminamos columna no at√≥mica
                        variables.remove(var)
                        df.drop(var, axis=1, inplace=True)
            
            # Se reemplazan los nan de los valores nuevos
            df.replace('nan', np.nan, inplace=True)

            variables = variables + variables_nuevas

            # T√©cnicas para tratar valores nulos
            # IMPORTANTE: Se asume que se debe implementar las t√©cnicas dropna y algun m√©todo de imputaci√≥n, y dar para elegir entre aplicar dropna o imputar valores
            if tecnica == "dropna":
                df.dropna(subset=variables, inplace=True)
            elif tecnica == "imputer":
                for var in variables:
                    data = df[var]
                    var_type = data.dtype
                    if is_numeric_dtype(var_type):
                        imputer = SimpleImputer(strategy="mean")
                        df[var] = imputer.fit_transform(df[var])
                    elif is_object_dtype(var_type):
                        imputer = SimpleImputer(strategy="most_frequent")
                        df[var] = SimpleImputer.fit_transform(df[var])
            else:
                raise ValueError("T√©cnica inv√°lida")
            
            return df
        
        # Se guarda la funci√≥n anterior en un FunctionTransformer y se ejecuta
        tranformer = FunctionTransformer(limpieza)
        self.ft = FunctionTransformer(limpieza)
        data_limpio = tranformer.transform(self.data)

        # Guardamos los datos en un archivo .csv
        data_limpio.to_csv(os.path.join(clean_dir, "data.csv"), index=False)

    def scale(self, tecnica="onehotencoder"):
        # Creaci√≥n de carpeta
        scale_dir = os.path.join(self.dir_fecha, "scale")
        os.makedirs(scale_dir, exist_ok=True)

        # Procesamiento de datos
        columnas_numericas = self.data.select_dtypes(include=['float', 'int']).columns
        columnas_categoricas = self.data.select_dtypes(exclude=['float', 'int']).columns

        transformers = []

        if columnas_numericas.any():
            class LogMinMax(BaseEstimator, TransformerMixin):
                    def fit(self, X, y=None):
                        return self

                    def transform(self, X):
                        X = np.log(X)
                        return (X - X.min()) / (X.max() - X.min())
            transformers.append(('log', LogMinMax(), columnas_numericas))

        # FALTA CORREGIR
        if columnas_categoricas.any():
            if tecnica == "onehotencoder":
                transformers.append(('onehotencoder', OneHotEncoder(), columnas_categoricas))
            if tecnica == "labelencoder":
                transformers.append(('labelencoder', LabelEncoder(), columnas_categoricas))

        ct = ColumnTransformer(transformers)
        self.ct = ColumnTransformer(transformers)
        columnas_transformadas = ct.fit_transform(self.data)

        columnas_transformadas_df = pd.DataFrame(columnas_transformadas, columns=list(columnas_numericas) + list(columnas_categoricas))
        # # Guardamos los datos en un archivo .csv
        columnas_transformadas_df.to_csv(os.path.join(scale_dir, "scaled_features.csv"), index=False)

    def make_clusters(self, metodo="kmeans"):
        # Creaci√≥n de carpeta
        clusters_dir = os.path.join(self.dir_fecha, "clusters")
        os.makedirs(clusters_dir, exist_ok=True)

        if metodo == "kmeans":
            pipe = Pipeline([('preprocesamiento', self.ft),
                            ('escalado', self.ct)])
            
            datos_limpios = pipe.fit_transform(self.data)

            # M√©todo del codo
            intertias = [
            [i, KMeans(n_clusters=i, random_state=42).fit(datos_limpios).inertia_]
            for i in range(2, 20)
            ]

            intertias = pd.DataFrame(intertias, columns=["n¬∞ clusters", "inertia"])

            # Se analiza el m√©todo del codo viendo el valor m√°ximo de la segunda derivada de la funci√≥n f(cluster)=inertia
            def segunda_derivada(df):
                lista = [np.nan]
                for i in np.arange(1, len(df) - 1):
                    lista.append(np.abs(df[i+1] + df[i-1] - 2*df[i]))
                lista.append(np.nan)
                return lista

            intertias['segunda_derivada'] = segunda_derivada(intertias)

            n_clusters = intertias[intertias['segunda_derivada'] == intertias['segunda_derivada'].max()]['n¬∞ clusters']
            cluster = KMeans(n_clusters=n_clusters, random_state=42)

        elif metodo == "dbscan":
            cluster = DBSCAN(eps=0.4, min_samples=3, random_state=42)

        tsne = TSNE(n_components=2, random_state=42)
        
        pipeline = Pipeline([('preprocesamiento', self.ft),
                            ('escalado', self.ct),
                            ('cluster', cluster),
                            ('tsne', tsne)])

        data = pipeline.fit_transform(self.data)


    def detect_anomalies(self):
        pass


    def profile(self, variable):
        self.summarize(variable)
        self.plot_vars
        self.clean_data
        self.scale
        self.make_clusters
        self.detect_anomalies




    def clearGarbage(self):
        dir_actual=os.getcwd()

        for carpeta in os.listdir(dir_actual):
          ruta_carpeta = os.path.join(dir_actual, carpeta)
          if os.path.isdir(ruta_carpeta):
             shutil.rmtree(ruta_carpeta)



            



    
```
